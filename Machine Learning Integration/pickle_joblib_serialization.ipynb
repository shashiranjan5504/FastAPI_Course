##Imports


```python
import pickle
import joblib
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

##Read the data



```python
df=pd.read_csv('/content/housing.csv').iloc[:,:-1].dropna()
df
```

```python

X = df.drop(columns='median_house_value')
y = df.median_house_value.copy()

print(X.shape, y.shape)
```

##Train the Model


```python
model = LinearRegression()
model.fit(X, y)
```

```python

y_pred = model.predict(X)
y_pred
```

it may be the formula of calculating the  model data

```python
r2_score(y, y_pred)
```

```python
model
```

yha tak hmara model train ho gya hai

##Pickle


code of serialization

```python
with open('model.pkl','wb') as f:
  pickle.dump(model,f)
```

deserialization

```python
with open('model.pkl','rb') as f:
  pickle_model=pickle.load(f)
```

now using the model_pickel trained model once again

```python
pickle_model
```

```python
y_pred_pickle = pickle_model.predict(X)
y_pred_pickle
```

```python
r2_score(y, y_pred_pickle)
```

##joblib


code of serialization

```python
joblib.dump(model,'model.joblib')
```

code of deseriliazation

```python
joblib_model=joblib.load('model.joblib')
joblib_model
```

```python
y_pred_joblib = joblib_model.predict(X)
y_pred_joblib
```

```python
r2_score(y, y_pred_joblib)
```

the output from both method comes equal
